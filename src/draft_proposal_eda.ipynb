### Average heart rate among Male and Female

```python
df_male = df.query('sex == 1')
df_male_avg = df_male['max heart rate'].mean()

df_female = df.query('sex == 0')
df_female_avg = df_female['max heart rate'].mean()

data = {'Sex': ['Male', 'Female'], 'Average Heart Rate': [df_male_avg, df_female_avg]}
df_heartrate = pd.DataFrame(data)
print(df_heartrate)
```

```python
fig = px.pie(df_heartrate, values='Average Heart Rate', names='Sex', title='Average Heart Rate Among Male and Female')
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

### Visualization of the recorded maximum heart rate among both the people affected with heart diseases and without heart diseases.

```python
fig = px.bar(
    df,
    x="target",
    y="max heart rate",
    color="target"
)
fig.update_xaxes(categoryorder="total descending")
fig.show()
```
